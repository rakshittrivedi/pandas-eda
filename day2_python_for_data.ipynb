````python
# ── Setup: paired data (we'll use dicts properly soon) ─────────
names  = ['Rakshit', 'Priya', 'Aman', 'Sneha', 'Rohan', 'Divya']
marks  = [88,        75,      92,     60,      45,      95     ]
branch = ['CS',      'EC',    'CS',   'ME',    'EC',    'CS'   ]

# SELECT name WHERE marks > 70
high_scorers = [names[i] for i in range(len(names)) if marks[i] > 70]
print("High scorers:", high_scorers)

# SELECT marks WHERE branch = 'CS'  (like SQL filter)
cs_marks = [marks[i] for i in range(len(marks)) if branch[i] == 'CS']
print("CS marks:", cs_marks)
print("CS average:", sum(cs_marks) / len(cs_marks))

# SELECT name, marks (transform — like SQL computed column)
grade_labels = [
    f"{names[i]}: {'PASS' if marks[i] >= 60 else 'FAIL'}"
    for i in range(len(names))
]
for label in grade_labels:
    print(label)
````

````python
# ── Utility Function 1: Standardise a text column ──────────────
def clean_text(value, default='UNKNOWN'):
    """
    Cleans a single text value:
    - Strips whitespace
    - Converts to uppercase
    - Replaces empty strings / None with a default

    Like SQL: UPPER(TRIM(COALESCE(value, 'UNKNOWN')))
    """
    if value is None:
        return default
    cleaned = str(value).strip().upper()
    return cleaned if cleaned else default

# Test it
test_vals = ['  CS ', 'ec', None, '', '  Me ']
for v in test_vals:
    print(f"  {repr(v):15} → {repr(clean_text(v))}")
````

````python
# ── Utility Function 2: Safe numeric converter ─────────────────
def to_number(value, default=0.0):
    """
    Safely converts any value to a float.
    Returns default if conversion fails (bad data, None, 'N/A', etc.)

    Like SQL: CAST(value AS FLOAT) — but without crashing on bad data.
    In SQL you'd need CASE WHEN isnumeric(val) THEN CAST... ELSE 0 END
    """
    try:
        return float(str(value).replace(',', '').strip())
    except (ValueError, TypeError):
        return default

# Test it — notice messy real-world data types
test_cases = [88, '75', '  92.5 ', '1,000', 'N/A', None, '', 'abc']
for t in test_cases:
    print(f"  {repr(t):12} → {to_number(t)}")
````

````python
# ── Utility Function 3: Categorise a numeric column ────────────
def grade_category(marks, thresholds=None):
    """
    Converts a numeric marks value into a grade label.

    Like SQL CASE WHEN marks >= 90 THEN 'A' WHEN marks >= 75 THEN 'B' ...
    thresholds is a list of (min_score, label) tuples, highest first.
    """
    if thresholds is None:
        thresholds = [
            (90, 'A — Distinction'),
            (75, 'B — First Class'),
            (60, 'C — Second Class'),
            (0,  'D — Pass'),
        ]
    cleaned = to_number(marks, default=-1)
    if cleaned < 0:
        return 'INVALID'
    for min_score, label in thresholds:
        if cleaned >= min_score:
            return label
    return 'FAIL'

# Apply to an entire list — one-liner using list comp
marks_list = [95, 82, 67, 55, 'N/A', None, 73, 91]
graded = [(m, grade_category(m)) for m in marks_list]
for marks, grade in graded:
    print(f"  {str(marks):8} → {grade}")
````

````python
# ── Combining all three: clean an entire "table" ───────────────
raw_students = [
    {'name': '  rakshit ', 'branch': 'cs',  'marks': '88'},
    {'name': 'PRIYA',      'branch': 'EC',  'marks': '75.0'},
    {'name': None,         'branch': None,  'marks': 'N/A'},
    {'name': '  Aman  ',   'branch': ' CS', 'marks': '92'},
]

def clean_student_row(row):
    """Full pipeline: clean one student dict."""
    return {
        'name':   clean_text(row.get('name'), default='UNNAMED'),
        'branch': clean_text(row.get('branch'), default='UNKNOWN'),
        'marks':  to_number(row.get('marks'), default=0.0),
        'grade':  grade_category(row.get('marks'))
    }

cleaned = [clean_student_row(r) for r in raw_students]
for r in cleaned:
    print(r)
````

````python
# ── Basic groupby: one column, one aggregation ─────────────────
# SQL: SELECT dept_name, AVG(salary) FROM df GROUP BY dept_name

dept_avg_salary = (df
    .groupby('dept_name')['salary']
    .mean()
    .round(0)
    .reset_index()
    .rename(columns={'salary': 'avg_salary'})
    .sort_values('avg_salary', ascending=False))

print("Average salary by department:")
print(dept_avg_salary)
````

````python
# ── Multi-column groupby with multiple aggregations ─────────────
# SQL: SELECT dept_name, gender, COUNT(*), AVG(salary), MAX(salary)
#      FROM df GROUP BY dept_name, gender

dept_gender = (df
    .groupby(['dept_name', 'gender'])
    .agg(
        headcount    = ('emp_id',   'count'),
        avg_salary   = ('salary',   'mean'),
        max_salary   = ('salary',   'max'),
        avg_rating   = ('rating',   'mean'),
        avg_exp      = ('experience','mean')
    )
    .round(2)
    .reset_index()
    .sort_values(['dept_name', 'headcount'], ascending=[True, False]))

print("\nDepartment × Gender breakdown:")
print(dept_gender)
````

````python
# ── Named aggregation — different agg per column ────────────────
city_summary = (df
    .groupby('city')
    .agg(
        employees    = ('emp_id',    'count'),
        total_salary = ('salary',    'sum'),
        avg_salary   = ('salary',    'mean'),
        avg_rating   = ('rating',    'mean'),
        min_exp      = ('experience','min'),
        max_exp      = ('experience','max'),
    )
    .round(2)
    .reset_index()
    .sort_values('total_salary', ascending=False))

print("\nCity-level HR summary:")
print(city_summary)
````

````python
# ── Custom aggregation with your own function ───────────────────
def salary_range(series):
    """Max minus min — spread of salaries in the group."""
    return series.max() - series.min()

dept_custom = (df
    .groupby('dept_name')
    .agg(
        salary_spread = ('salary', salary_range),
        top_performer = ('avg_score', 'max'),
    )
    .reset_index())

print("\nCustom aggregations:")
print(dept_custom)
````

````python
# ── transform() — like window functions, keeps all rows ─────────
# SQL: AVG(salary) OVER (PARTITION BY dept_name)
# Every row gets its department's average, not collapsed

df['dept_avg_salary']   = df.groupby('dept_name')['salary'].transform('mean')
df['salary_vs_dept']    = (df['salary'] - df['dept_avg_salary']).round(0)
df['dept_rank_salary']  = (df.groupby('dept_name')['salary']
                             .rank(method='dense', ascending=False)
                             .astype('Int64')) # Changed to nullable integer type

print("\nWindow-function-style transform:")
print(df[['full_name','dept_name','salary',
          'dept_avg_salary','salary_vs_dept','dept_rank_salary']].head(8))
````

````python
# ── Column-level apply (most common) ───────────────────────────

# 1. Clean name — strip + title case
df['name_clean'] = df['full_name'].apply(
    lambda x: x.strip().title() if pd.notna(x) else 'Unknown'
)

# 2. Salary band classification
def salary_band(sal):
    if pd.isna(sal):   return 'Unknown'
    if sal >= 90000:   return 'Senior'
    if sal >= 70000:   return 'Mid-level'
    if sal >= 50000:   return 'Junior'
    return 'Intern'

df['salary_band'] = df['salary'].apply(salary_band)

# 3. Experience tier
df['exp_tier'] = df['experience'].apply(
    lambda x: (
        'Expert'    if pd.notna(x) and x >= 6 else
        'Mid'       if pd.notna(x) and x >= 3 else
        'Fresher'   if pd.notna(x)             else
        'Unknown'
    )
)

print(df[['name_clean','salary','salary_band','experience','exp_tier']].head(8))
````

````python
# ── Row-level apply (axis=1) — access multiple columns per row ──
# Use when your transformation needs more than one column at once

def performance_flag(row):
    """
    Flag employees needing attention:
    Low rating AND below-dept-average salary = 'At Risk'
    Low rating but fair salary               = 'Needs Coaching'
    Good rating                              = 'On Track'
    No rating data                           = 'Review Pending'
    """
    if pd.isna(row['rating']):
        return 'Review Pending'
    if row['rating'] < 3.5 and row['salary_vs_dept'] < -5000:
        return 'At Risk'
    if row['rating'] < 3.5:
        return 'Needs Coaching'
    return 'On Track'

df['perf_flag'] = df.apply(performance_flag, axis=1)

print("\nPerformance flags:")
print(df[['name_clean','rating','salary_vs_dept','perf_flag']].head(10))

# Distribution of flags
print("\nFlag distribution:")
print(df['perf_flag'].value_counts())
````

````python
# ── applymap/map — element-wise on entire DataFrame ─────────────
# Useful for formatting a whole table for export

numeric_cols = ['salary', 'dept_avg_salary', 'budget']
display_df = df[numeric_cols].copy()

# Format every number as currency string (for reporting)
display_df = display_df.map(
    lambda x: f"₹{x:,.0f}" if pd.notna(x) else "N/A"
)
print("\nFormatted for display:")
print(display_df.head(4))
````

````python
# ── Step 1: Audit every column for nulls ───────────────────────
print("NULL AUDIT")
print("=" * 40)
null_report = pd.DataFrame({
    'null_count':   df.isna().sum(),
    'null_pct':     (df.isna().mean() * 100).round(1),
    'dtype':        df.dtypes
})
print(null_report[null_report['null_count'] > 0])
````

````python
# ── Step 2: fillna strategies ───────────────────────────────────

df_clean = df.copy()   # always work on a copy, keep original safe

# Strategy 1: Fill with a fixed value (for categorical)
df_clean['dept_name']  = df_clean['dept_name'].fillna('Unassigned')
df_clean['email']      = df_clean['email'].fillna('noemail@techcorp.com')
df_clean['name_clean'] = df_clean['name_clean'].fillna('Unknown Employee')

# Strategy 2: Fill numeric with column MEDIAN (robust to outliers)
# (don't use mean if you have outliers — that 450000 salary will skew it)
median_exp    = df_clean['experience'].median()
median_rating = df_clean['rating'].median()

df_clean['experience'] = df_clean['experience'].fillna(median_exp)
df_clean['rating']     = df_clean['rating'].fillna(median_rating)
df_clean['avg_score']  = df_clean['avg_score'].fillna(df_clean['avg_score'].median())

# Strategy 3: Forward fill — for time-series / ordered data
# df.sort_values('date')['metric'].fillna(method='ffill')

# Strategy 4: Group-wise fill — fill with GROUP median
# (better than overall median — use department's median, not company's)
df_clean['salary'] = df_clean.groupby('dept_name')['salary'].transform(
    lambda x: x.fillna(x.median())
)

print("Nulls after cleaning:")
print(df_clean[['experience','rating','avg_score','salary']].isna().sum())
````

````python
# ── When to dropna ──────────────────────────────────────────────
# Rule: only drop if the row is genuinely unusable

before = len(df_clean)

# Drop only if BOTH name AND dept are missing (truly ghost records)
df_clean = df_clean.dropna(subset=['name_clean'], how='all')

after = len(df_clean)
print(f"Dropped {before - after} rows | {after} remain")
````

````python
# ── Step 3: Fix data types ──────────────────────────────────────
print("BEFORE dtype fixes:")
print(df_clean[['dept_id','join_year','salary','rating']].dtypes)

# Fix join_year — it's a string, has one bad value 'twenty19'
def safe_year(val):
    try:
        year = int(str(val).strip())
        return year if 2000 <= year <= 2025 else None  # sanity check
    except (ValueError, TypeError):
        return None   # 'twenty19' → None → we'll fill it next

df_clean['join_year'] = df_clean['join_year'].apply(safe_year)
df_clean['join_year'] = df_clean['join_year'].fillna(
    df_clean['join_year'].median()
).astype(int)

# Fix dept_id — should be int, not float (NaN forces float in Pandas)
df_clean['dept_id'] = df_clean['dept_id'].fillna(0).astype(int)

# Downcast salary to int (no decimals in salary)
df_clean['salary'] = df_clean['salary'].round(0).astype(int)

# Verify
print("\nAFTER dtype fixes:")
print(df_clean[['dept_id','join_year','salary','rating']].dtypes)
````

````python
# ── The IQR method — the industry standard ─────────────────────
#
# IQR = Q3 - Q1  (the "middle 50%" spread)
# Lower fence = Q1 - 1.5 × IQR
# Upper fence = Q3 + 1.5 × IQR
# Anything outside the fences = outlier
#
# Why 1.5? Statistician John Tukey's rule — captures ~99.3% of
# normal data. Use 3.0 for "extreme outliers only" mode.
#
# In data engineering: outliers aren't always errors —
# they're flags for investigation. Don't auto-delete them.

def detect_outliers_iqr(df, column, threshold=1.5):
    """
    Detects outliers in a numeric column using the IQR method.

    Returns a dict with:
    - outlier_mask   : boolean Series (True = outlier row)
    - outlier_rows   : DataFrame of outlier rows
    - bounds         : (lower_fence, upper_fence)
    - stats          : Q1, Q3, IQR values
    """
    col = df[column].dropna()

    Q1  = col.quantile(0.25)
    Q3  = col.quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - threshold * IQR
    upper = Q3 + threshold * IQR

    outlier_mask = (df[column] < lower) | (df[column] > upper)

    return {
        'outlier_mask': outlier_mask,
        'outlier_rows': df[outlier_mask][['name_clean', column]],
        'bounds':       (round(lower, 2), round(upper, 2)),
        'stats':        {'Q1': Q1, 'Q3': Q3, 'IQR': IQR},
        'count':        outlier_mask.sum()
    }


# ── Detect salary outliers ──────────────────────────────────────
salary_result = detect_outliers_iqr(df_clean, 'salary')

print("SALARY OUTLIER REPORT")
print("=" * 40)
print(f"Q1:          ₹{salary_result['stats']['Q1']:>10,.0f}")
print(f"Q3:          ₹{salary_result['stats']['Q3']:>10,.0f}")
print(f"IQR:         ₹{salary_result['stats']['IQR']:>10,.0f}")
print(f"Lower fence: ₹{salary_result['bounds'][0]:>10,.0f}")
print(f"Upper fence: ₹{salary_result['bounds'][1]:>10,.0f}")
print(f"Outliers found: {salary_result['count']}")
print()
print(salary_result['outlier_rows'])
````

````python
# ── Scan ALL numeric columns at once ───────────────────────────
numeric_cols = ['salary', 'rating', 'experience', 'avg_score', 'total_tasks']

print("OUTLIER SCAN — ALL NUMERIC COLUMNS")
print("=" * 55)

outlier_summary = []
for col in numeric_cols:
    if col in df_clean.columns:
        result = detect_outliers_iqr(df_clean, col)
        outlier_summary.append({
            'column':      col,
            'outliers':    result['count'],
            'lower_fence': result['bounds'][0],
            'upper_fence': result['bounds'][1],
        })
        if result['count'] > 0:
            print(f"\n⚠️  {col}: {result['count']} outlier(s)")
            print(result['outlier_rows'].to_string(index=False))

print("\n\nSUMMARY TABLE:")
print(pd.DataFrame(outlier_summary).to_string(index=False))
````

````python
# ── Three strategies for handling outliers ─────────────────────

# Strategy 1: CAP (winsorize) — replace with fence value
# Best for: keeping all rows, just limiting extreme influence
def cap_outliers(df, column, threshold=1.5):
    result = detect_outliers_iqr(df, column, threshold)
    lower, upper = result['bounds']
    df[column] = df[column].clip(lower=lower, upper=upper)
    return df

# Strategy 2: FLAG — add a boolean column, keep the data
# Best for: you want to analyse outliers separately
salary_flags = detect_outliers_iqr(df_clean, 'salary')
df_clean['salary_outlier'] = salary_flags['outlier_mask']

# Strategy 3: REMOVE — only when the value is definitively wrong
# (e.g. a salary of 450000 when max budget per person is 200000)
# df_clean = df_clean[~salary_flags['outlier_mask']]  # uncomment to drop

# We'll use cap for salary (keeps data, removes extreme skew)
df_clean = cap_outliers(df_clean, 'salary')

print("After capping salary outliers:")
print(f"  Max salary: ₹{df_clean['salary'].max():,.0f}")
print(f"  Min salary: ₹{df_clean['salary'].min():,.0f}")
````

````python
# ── Final cleanup before saving ─────────────────────────────────

# Select and order final columns
final_cols = [
    'emp_id', 'name_clean', 'dept_id', 'dept_name', 'city',
    'salary', 'salary_band', 'salary_vs_dept', 'dept_rank_salary',
    'join_year', 'experience', 'exp_tier',
    'rating', 'avg_score', 'total_tasks',
    'perf_flag', 'salary_outlier', 'gender', 'email'
]

# Keep only columns that exist (safety check)
final_cols = [c for c in final_cols if c in df_clean.columns]
df_final   = df_clean[final_cols].copy()

print("FINAL CLEAN DATASET")
print(f"Shape: {df_final.shape}")
print()
df_final.info()
````

````python
# ── Save to SQLite database ─────────────────────────────────────
# SQLite = a real database stored as a single file (.db)
# Perfect for portfolio projects — no server needed

db_path = 'techcorp_hr.db'
conn    = sqlite3.connect(db_path)    # creates file if it doesn't exist

# Write employees table
df_final.to_sql(
    name='employees',       # table name in the DB
    con=conn,
    if_exists='replace',    # 'replace' = drop & recreate | 'append' = add rows
    index=False
)

# Write departments as a separate table too
departments_raw.to_sql(
    name='departments',
    con=conn,
    if_exists='replace',
    index=False
)

# Write performance as a separate table
performance_raw.to_sql(
    name='performance',
    con=conn,
    if_exists='replace',
    index=False
)

print(f"✅ SQLite DB saved: {db_path}")

# Verify: list all tables in the DB
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", conn)
print(f"   Tables in DB: {tables['name'].tolist()}")
````

````python
# ── Query your SQLite DB with SQL — full circle moment ──────────
# You just cleaned data in Python, stored it in SQL,
# and now you're querying it with SQL. This is the pipeline.

print("QUERYING THE CLEAN DB WITH SQL")
print("=" * 45)

# Query 1: Department summary
q1 = pd.read_sql("""
    SELECT
        dept_name,
        COUNT(*)                    AS headcount,
        ROUND(AVG(salary), 0)       AS avg_salary,
        ROUND(AVG(rating), 2)       AS avg_rating
    FROM employees
    GROUP BY dept_name
    ORDER BY avg_salary DESC
""", conn)
print("\nDepartment Summary:")
print(q1)

# Query 2: Performance flags
q2 = pd.read_sql("""
    SELECT perf_flag, COUNT(*) AS count
    FROM employees
    GROUP BY perf_flag
    ORDER BY count DESC
""", conn)
print("\nPerformance Flag Distribution:")
print(q2)

# Query 3: Join employees with performance data
q3 = pd.read_sql("""
    SELECT
        e.name_clean,
        e.dept_name,
        e.salary,
        AVG(p.score)      AS avg_monthly_score,
        SUM(p.tasks_done) AS total_tasks
    FROM employees e
    LEFT JOIN performance p ON e.emp_id = p.emp_id
    GROUP BY e.emp_id, e.name_clean, e.dept_name, e.salary
    ORDER BY avg_monthly_score DESC
    LIMIT 5
""", conn)
print("\nTop 5 by Performance Score:")
print(q3)

conn.close()
print("\n✅ Connection closed")
````
